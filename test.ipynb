```python
forecast = result.get_prediction(start=-10, dynamic=True)

mean_forecast = forecast.predicted_mean

plt.plot(apple_df[-100:], label='Historical Returns')
plt.plot(mean_forecast.index, mean_forecast, label='In Sample Forecast', color='orange')
plt.title('ARIMA Mean Forecast (Last 10 Days of Apple Returns)')
plt.legend()
plt.show()
```

```python
# Assume `returns` is your series and `result` is the fitted ARIMA model
last_10_idx = apple_df.index[-20]  # get the timestamp of the 10th-to-last observation

forecast = result.get_forecast(steps=10)
mean_forecast = forecast.predicted_mean

# Plot last 100 days of actual returns
plt.plot(apple_df[-100:], label='Historical Returns')

# Plot forecast, aligning properly
plt.plot(mean_forecast.index, mean_forecast, label='Forecast', color='orange')

plt.title('ARIMA Mean Forecast (Next 10 Days of Apple Returns)')
plt.legend()
plt.show()
```

```python
import torch
import torch.nn as nn

class StockMLP(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.net = nn.Sequential(
            nn.Flatten(),                        # shape: (batch, 10, 6) → (batch, 60)
            nn.Linear(10 * num_features, 128),   # input → 128 neurons
            nn.ReLU(),
            nn.Linear(128, 64),                  # 128 → 64 neurons
            nn.ReLU(),
            nn.Linear(64, 1)                    # final output: 10-day prediction
        )

    def forward(self, x):
        return self.net(x)
```

```python
import torch.nn as nn

class StockNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.network = nn.Sequential(
            nn.Flatten(),            # (10, 6) → (60,)
            nn.Linear(60, 64),       
            nn.ReLU(),
            nn.Linear(64, 10)         # Output: 1 value
        )

    def forward(self, x):
        return self.network(x)

model = StockNN()
```

```python
import torch.optim as optim

loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# Put model in training mode
model.train()

for epoch in range(10):  # train for 10 epochs
    total_loss = 0

    for batch_x, batch_y in train_loader:
        print(batch_x.size())
        optimizer.zero_grad()          # clear previous gradients
        pred = model(batch_x)          # forward pass
        loss = loss_fn(pred, batch_y)  # compute loss
        loss.backward()                # backpropagation
        optimizer.step()               # update weights

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
import os
import pandas as pd
dir_path = 'data/train/stocks/'
stocks_lst = []

for stock in sorted(os.listdir(dir_path)): 
    file_path = os.path.join(dir_path, stock)
    df = pd.read_csv(file_path)
    stocks_lst.append(df)

stocks_df = pd.concat(stocks_lst)

print(stocks_df) 
```

```python
sp500_df = pd.read_csv('data/train/indices/SP500.csv')
sp500_df = sp500_df.rename(columns={'Returns': 'sp500_return'})
sp500_df = sp500_df[['Date', 'sp500_return']]

print(sp500_df)
```

```python
merged_df = pd.merge(stocks_df, sp500_df, on='Date')
print(merged_df)
merged_df = merged_df.sort_values(by=['Date', 'Ticker'])

print(merged_df)
```

```python
def compute_rolling_beta(group, window=60):
    cov = group['Returns'].rolling(window).cov(group['sp500_return'])
    var = group['sp500_return'].rolling(window).var()
    return cov / var

merged_df['rolling_beta'] = (
    merged_df.groupby('Ticker', group_keys=False)
    .apply(compute_rolling_beta)
)

print(merged_df)
```

```python
import pandas as pd 
from src.linear_model import train, test, predict
from src.new_dataset import LinearDataset
from src.features import beta_features

#Train the model 
def main():
    
    dataset = LinearDataset("data/train/stocks/")
    X, Y = dataset.preprocess()
    X_train, X_test, Y_train, Y_test = dataset.split(X, Y)

    model = train(X_train, Y_train)
    
    accuracy, pred = test(model, X_test, Y_test)
    print(f"The MSE of the model is {accuracy}.")
    predictions = []
    for i in range(1, 6):
        test_path = f"data/test/test_{i}.csv"
        df = pd.read_csv(test_path)
        df = beta_features(stocks_dir=test_path, sp500_path="data/train/indices/SP500.csv")

        X_pred = df[["Open", "High", "Low", "Close", "Adjusted", "Volume", "beta_lag1"]].values
        X_pred = X_pred[:-10] 

        Y_pred = predict(model, X_pred)
        predictions.append(Y_pred[:10].flatten())

    pred_matrix = list(zip(*predictions))

    # Create submission DataFrame
    dates = ["2025-03-25", "2025-03-26", "2025-03-27", "2025-03-28", "2025-03-31", "2025-04-01", "2025-04-02", "2025-04-03", "2025-04-04", "2025-04-07"]
    submission_df = pd.DataFrame(pred_matrix, columns=["Returns_1", "Returns_2", "Returns_3", "Returns_4", "Returns_5"])
    submission_df.insert(0, "Date", dates)

    # Save to CSV
    submission_df.to_csv("submission.csv", index=False)
    print("Saved submission.csv!")

if __name__ == "__main__":
    main()

```

```python
import torch.nn as nn

class StockNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Flatten(),             # turns (10, 6) into (60,)
            nn.Linear(60, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, 10)        # output: 10 predicted returns
        )

    def forward(self, x):
        return self.model(x)
```

```python
model = StockNN()
print(model)
```

```python
loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
model.train()  # set model to training mode

for batch_x, batch_y in train_loader:
    optimizer.zero_grad()         # clear gradients
    pred = model(batch_x)         # forward pass
    loss = loss_fn(pred, batch_y) # calculate loss
    loss.backward()               # backpropagation
    optimizer.step()              # update weights
    print("Loss:", loss.item())
    break  # ⛔ Just run 1 batch for now to test logic
```

```python
epochs = 10

for epoch in range(epochs):
    total_loss = 0

    for batch_x, batch_y in train_loader:
        optimizer.zero_grad()
        pred = model(batch_x)
        loss = loss_fn(pred, batch_y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    print(f"📉 Epoch {epoch+1} | Avg Loss: {total_loss / len(train_loader):.6f}")
```
